```python
df=pd.read_csv("Melbourne_housing_FULL.csv")
df
```

```python
df1=df.bfill()
df1.ffill()
```

```python
df1.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df1['Suburb']= le.fit_transform(df1['Suburb'])
df1['Address']= le.fit_transform(df1['Address'])
df1['Type']=le.fit_transform(df1['Type'])
df1['Method']=le.fit_transform(df1['Method'])
df1['SellerG']=le.fit_transform(df1['SellerG'])
df1['CouncilArea']=le.fit_transform(df1['CouncilArea'])
df1['Regionname']=le.fit_transform(df1['Regionname'])
```

```python
df1
```

```python
df2=df1.drop('Date', axis=1)
```

```python
df2.corr()
```

```python
x=df2.drop("Price",axis=1)
y=df2['Price']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.2)
```

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
xtrain=sc.fit_transform(xtrain)
xtest=sc.fit_transform(xtest)
```

```python
x.shape
```

```python
ann=tf.keras.models.Sequential()
```

```python
ann.add(tf.keras.layers.Dense(units=200, input_dim=19, activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense(units=200,activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense(units=1, activation='linear'))
```

```python
ann.compile(optimizer='adam',loss='mean_squared_error',metrics=tf.keras.metrics.RootMeanSquaredError())
```

```python
ann.fit(xtrain,ytrain,batch_size=30, epochs=30)
```

```python
ann.evaluate(xtest,ytest)
```

```python
pred=ann.predict(xtest)
print(pred)
```

```python
print(ytest)
```

```python
from sklearn.metrics import r2_score
acc=r2_score(ytest, pred)
acc
```
